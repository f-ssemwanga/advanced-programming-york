## Exercise one
Write a program in Python that asks the user for a sentence and analyses it to establish the number of separate vowels (A, E, I, O, U) and the number of “other” non-space characters contained in the sentence. Your program should output a histogram which indicates a * (star) for each vowel character, followed by a statement indicating the number of other non-space characters.

For example, if the input was: 

**The black cat sat up on the orange mat!**

The output would be:
![Exercise 1 Output](images/functionsExercise1.png)

```python
def sentenceAnalyser():
  '''takes a sentence and outputs an analysis histogram'''
  sentence = input('Enter a sentence to analyse: ')
  ast ="*"
  aTotal,eTotal,iTotal,oTotal,uTotal,otherTotal = (0,0,0,0,0,0)
  splitSentence = sentence.split(" ")
  print(splitSentence)
  for i in range(len(splitSentence)):
    for char in splitSentence[i]:
      if char.lower() =="a":
        aTotal +=1
      elif char.lower() =="e":
        eTotal +=1
      elif char.lower() =="i":
        iTotal +=1
      elif char.lower() =="o":
        oTotal +=1
      elif char.lower() =="u":
        uTotal +=1
      elif char.lower() !=" ":
        otherTotal +=1
  
  print(f'A:{aTotal}|E:{eTotal}|I:{iTotal}|O:{oTotal}|U:{uTotal}|other:{otherTotal}')
  print(f'A:\t{ast*aTotal}')
  print(f'E:\t{ast*eTotal}')
  print(f'I:\t{ast*iTotal}')
  print(f'O:\t{ast*oTotal}')
  print(f'U:\t{ast*uTotal}')
  print(f'Other (non-space) Characters:\t{otherTotal}')
```

```python
sentenceAnalyser()
```

## Exercise two
Write a program in Python that checks if two inputted words are ‘anagrams’. If the user inputs ‘note’ and ‘tone’, the program would output “These are anagrams”. If the user entered 'cat' and 'sat' the program would output "These are not anagrams". Your program should contain a function “isAnagram()” that returns either true or false. The main part of the program should provide a suitable interface for the user. 

```python
def isAnagram(word1,word2):
  '''takes two word and returns whether they are anagrams or not'''
  wordsList1 = list(word1)
  wordsList2 = list(word2)
  if sorted(wordsList1) == sorted(wordsList2):
    return True
  else:
    return False
```

```python
def anagramAnalyser():
  '''takes two words from the users and outputs whether the words are anagrams or not'''
  words= []
  for i in range(2):
    word = input(f'Enter word{i+1}')
    words.append(word)
  if isAnagram(words[0],words[i]) == True:
    print('These are anagrams')
  else:
    print("These are not anagrams")
anagramAnalyser()
```

## Exercise three
Revisit the programming task for "Activity 4 Exercise 1" where you emulated a stack and queue using a list. Consider how these programs could be combined and how functions could be used to make the code more concise and reusable.

Write a new program to demonstrate this, making sure you offer the user the option of using a stack or a queue first and then applying the appropriate constraints.

```python
#stack demo application
def inputValidator(inputString):
  '''returns numeric user choice'''
  while True:
    try:
      choice = int(input(f"Enter a {inputString}:"))
      return choice
    except ValueError:
      print('Enter a numeric value')

def stackOperation():
  '''stack demo'''
  stackItems = []
  while True:
    print('-'*30)
    print('Choose Stack operations\n1. Push\n2. Pop\n3. View\n4. Return to main menu')
    operation =inputValidator("stack operation")
    print('-'*15)
    print()
    if operation ==1:
      print(f'Original Stack status: {stackItems}')
      toPush = inputValidator("value to push onto the stack")
      print(f'Original Stack: {stackItems}')
      stackItems.insert(0,toPush)
      print(f'Updated Stack status: {stackItems}')
      continue
    elif operation == 2:
      if len(stackItems) ==0:
        print('No items to pop! Stack is empty')
        continue
      else:
        print(f'Original Stack: {stackItems}')
        del stackItems[0]
        print(f'Updated Stack status: {stackItems}')
        continue
    elif operation ==3:
      print(f'Current Stack status: {stackItems}')
      continue
    elif operation == 4:
      break
    else:
      print('Invalid menu option')
def queueOperation():
  '''queue demo'''
  queueItems = []
  while True:
    print('-'*30)
    print('Choose Queue operation: \n1. Push\n2. Pop\n3. View\n4. Return to main menu')
    operation =inputValidator("queue operation")
    print('-'*30)
    print()
    if operation ==1:
      print(f'Original Queue status: {queueItems}')
      toPush = inputValidator("value to push onto the queue")
      print(f'Original Queue: {queueItems}')
      queueItems.append(toPush)
      print(f'Updated Queue status: {queueItems}')
      continue
    elif operation == 2:
      if len(queueItems) ==0:
        print('No items to pop! Queue is empty')
        continue
      else:
        print(f'Original Queue: {queueItems}')
        del queueItems[0]
        print(f'Updated Queue status: {queueItems}')
        continue
    elif operation ==3:
      print(f'Current Queue status: {queueItems}')
      continue
    elif operation == 4:
      break   
    else:
     print('Invalid menu option') 
  
```

```python
def main():
  '''Menu to demo a stack'''
  while True:
    print("Choose a structure to test:")
    print('1. Stack\n2. Queue\n3. Quit')
    userChoice = inputValidator("menu choice")
    if userChoice == 1:
      stackOperation()
    if userChoice == 2:
      queueOperation()
    elif userChoice == 3:
      quit()
```

```python
main()
```
